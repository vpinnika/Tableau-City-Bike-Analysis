```python
# Importing Dependencies
import pandas as pd
import csv
import time
import os
```

```python
#Defining source file path
apr19_file = os.path.join("Resources","JC-201904-citibike-tripdata.csv")
may19_file = os.path.join("Resources","JC-201905-citibike-tripdata.csv")
jun19_file = os.path.join("Resources","JC-201906-citibike-tripdata.csv")
jul19_file = os.path.join("Resources","JC-201907-citibike-tripdata.csv")
aug19_file = os.path.join("Resources","JC-201908-citibike-tripdata.csv")
sep19_file = os.path.join("Resources","JC-201909-citibike-tripdata.csv")
oct19_file = os.path.join("Resources","JC-201910-citibike-tripdata.csv")
nov19_file = os.path.join("Resources","JC-201911-citibike-tripdata.csv")
dec19_file = os.path.join("Resources","JC-201912-citibike-tripdata.csv")
jan20_file = os.path.join("Resources","JC-202001-citibike-tripdata.csv")
feb20_file = os.path.join("Resources","JC-202002-citibike-tripdata.csv")
mar20_file = os.path.join("Resources","JC-202003-citibike-tripdata.csv")
```

```python
#Reading the files into Pandas
apr19 = pd.read_csv(apr19_file)

may19 = pd.read_csv(may19_file)

jun19 = pd.read_csv(jun19_file)
                    
jul19 = pd.read_csv(jul19_file)
                    
aug19 = pd.read_csv(aug19_file)
                    
sep19 = pd.read_csv(sep19_file)
                    
oct19 = pd.read_csv(oct19_file)
                    
nov19 = pd.read_csv(nov19_file)
                    
dec19 = pd.read_csv(dec19_file)
                    
jan2020 = pd.read_csv(jan20_file)
                    
feb2020 = pd.read_csv(feb20_file)
                    
mar2020 = pd.read_csv(mar20_file)
```

```python
apr19.name = 'april'
may19.name = 'may'
jun19.name = 'june'
jul19.name = 'july'
aug19.name = 'august'
sep19.name = 'september'
oct19.name = 'october'
nov19.name = 'november'
dec19.name = 'december'
```

```python
mar2020.head()
```

```python
#Updating/adding columns 
dfs2019 = [apr19,may19,jun19,jul19,aug19,sep19,oct19,nov19,dec19]

for df in dfs2019:
    df['age'] = 2019-df['birth year']
    df['month'] = df.name
    df['year'] = 2019
    newGenList = []
    genList = list(df['gender'])
#     time.sleep(1)
    print(genList)
    for x in genList:
        if x == 1:
            newGenList.append('Male')
        elif x == 2:
            newGenList.append('Female')
        else:
            newGenList.append('Unknown')
    df['gender'] = ""
    df['gender'] = newGenList
    
```

```python
# Merging DFs
merge_2019_data = pd.concat(dfs2019)
merge_2019_data.head()
```

```python
#Updating/adding columns 
dfs2020 = [jan2020,feb2020,mar2020]

for df in dfs2020:
    df['age'] = 2020-df['birth year']
    newGenList = []
    genList = list(df['gender'])
    for x in genList:
        if x == 1:
            newGenList.append('Male')
        elif x == 2:
            newGenList.append('Female')
        else:
            newGenList.append('Unknown')
    df['gender'] = ""
    df['gender'] = newGenList
    
    
```

```python
jan2020['month'] = 'january'
feb2020['month'] = 'february'
mar2020['month'] = 'march'

jan2020['year'] = '2020'
feb2020['year'] = '2020'
mar2020['year'] = '2020'
```

```python
merge_2020_data = pd.concat(dfs2020)

merge_2020_data.head()
```

```python
final_data = pd.concat([merge_2019_data,merge_2020_data],ignore_index=True)
```
